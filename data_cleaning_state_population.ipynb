# Data cleaning for getting state population

```python
import pandas as pd
import json
```

```python
df = pd.read_csv("pop2019.csv",encoding = "ISO-8859-1")

# Get the conversion mapping to convert from State Name to State Code.
with open('state_conversion.json') as f:
  conv = json.load(f)
conv = dict(map(reversed, conv.items()))
# print(conv)

df["state"] = df["STNAME"].map(conv)
df["pop"] = df["POPESTIMATE2019"]
df = df[["state", "pop"]]

# Getting sum of all counties withi
df2 = df.groupby("state").sum()
```
