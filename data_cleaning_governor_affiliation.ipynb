## Data cleaning for party affiliation data

```python
import pandas as pd
```

```python
# Reading the downloaded dataset
df = pd.read_csv("governor_affiliation.csv", )
```

```python
df
```

```python
# Create new row to state if it is Democrat or not
def func(row):
    if row[1] == "D":
        return True
    return False

simple_df = df.copy()
simple_df["is_democrat"] =  simple_df.apply(func = func, axis=1)
```

```python
simple_df
```

```python
# Save to csv file
simple_df.to_csv("party_affiliation.csv")
```
