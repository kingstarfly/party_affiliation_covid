## Data cleaning for party affiliation data

```python
import pandas as pd
```

```python
# Reading the downloaded dataset
df = pd.read_csv("raw_governor_affiliation.csv")
```

```python
import json
with open("state_conversion.json") as f:
    conv = json.load(f)
    reversed_conv = {value : key for (key, value) in conv.items()}
    df["state"] = df["state"].replace(reversed_conv)
```

```python
df
```

```python
# Create new row to state if it is Democrat or not
def func(row):
    if row[1] == "D":
        return True
    return False

simple_df = df.copy()
simple_df["is_democrat"] =  simple_df.apply(func = func, axis=1)
```

```python
simple_df
```

```python
# Save to csv file
simple_df.to_csv("governor_affiliation.csv", index=False)
```
